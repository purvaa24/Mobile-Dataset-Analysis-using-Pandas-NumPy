Part 1: Data Loading & Initial Inspection

```python
#1. Reading Files:
import pandas as pd
import numpy as np
data=pd.read_csv("/content/Mobiles_Dataset.csv")
data.head(5)
```

```python
data.tail(3)
```

```python
#Basic Info:
df=pd.DataFrame(data)
df.info()
```

```python
df.count()
```

```python
#Descriptive Stats:
df.describe()
```

```python
#4. Sampling:
df.sample(10)
```

```python
df.rename(columns={'Actual price':'MRP','Discount price':'Selling_Price'},inplace=True)
df.head(5)
```

```python
#Part 2: Data Cleaning & Type Conversion
```

```python
#6. String Processing & astype:
df['MRP']=df['MRP'].astype(str).str.replace('₹','').str.replace(',','').replace('NIL',np.nan).astype(float)
df['MRP']
```

```python
df['Selling_Price']=pd.to_numeric(df['Selling_Price'].astype(str).str.replace('₹','').str.replace(',',''),errors='coerce')
df['Selling_Price']
```

```python
df['Rating']=df['Rating'].str.replace('Ratings','', regex=True).str.replace(',','').astype(int)
print(df['Rating'])
df['Reviews']=df['Reviews'].str.replace('Reviews','', regex=True).str.replace(',','').astype(int)
print("\n",df['Reviews'])
```

```python
#7. Handling Missing Data:
df['Camera'].isnull().sum()
```

```python
df['Camera'].fillna(value='Not Specified')
```

```python
df.dropna(how='any',subset=['Stars'])
```

```python
#8. Duplicates:
df.drop_duplicates()
```

```python
#9. Replace:
df['RAM (GB)']=df['RAM (GB)'].str.replace('NIL','np.nan')
df['RAM (GB)']
```

```python
#Part 3: Selection, Filtering & Querying
```

```python
#10. loc vs iloc:
a=df.iloc[0:10, 0:3]
print(a)
```

```python
b=df.loc[df['Product Name'].str.contains('iPhone'),['Product Name', 'Selling_Price']]
print(b)
```

```python
#11.Boolean Indexing & isin:
df['RAM (GB)'].isin(['8', '12'])
```

```python
(df['Stars'] > 4.5) & (df['Selling_Price'] < 50000)
```

```python
#12. query:
query=df.query('Product Name == SAMSUNG' and 'Selling_Price > 70000')
print(query)
```

```python
#13. where:
where=df.where(df['Selling_Price'] > 20000, 0)
print(where)
```

```python
#Part 4: Advanced Series & String Operations
```

```python
#14. value_counts & unique:
a=df['Product Name'].nunique()
print(a)
b=df['Product Name'].value_counts()
print(b)
```

```python
#15. map:
df['Price_Category'] = pd.cut(df['Stars'],bins=[1,3,4,5],labels=['Budget','Mid-Range','Premium'])
df.head(10)
```

```python
#16. apply & assign:
Discount_Percentage=df.apply(lambda x:((x['MRP'] -x['Selling_Price'])/x['MRP'])*100,axis=1)
print(Discount_Percentage)
```

```python
#17. std & mean:
std= df['Selling_Price'].std()
print(std)
mean= df['Selling_Price'].mean()
print(mean)
```

```python
#Part 5: Grouping & Aggregation
```

```python
#18. groupby & agg:
group=df.groupby('Product Name')
print(group[['Selling_Price','Stars']].agg(np.mean))
```

```python
print(group[['Selling_Price','Stars']].agg(np.sum))
```

```python
print(group[['Selling_Price','Stars']].agg(np.std))
```

```python
#19.transform:
df['Average_Brand']=df.groupby('Product Name')['Selling_Price'].transform(np.mean)
df.head(2)
```

```python
#20. pivot_table:
pivot = pd.pivot_table(df,values='Stars',index='Product Name',columns='RAM (GB)',aggfunc=np.mean)
print(pivot)
```

```python
#21. crosstab / count:
s= df.groupby('Product Name')['Storage (GB)'].count()
print(s)
```

```python
#Part 6: Reshaping & Combining
```

```python
#22.transpose & melt:
s=df.describe()
a=s.transpose()
print(a)
```

```python
pd.melt(df,id_vars=['Product Name'],value_vars=['MRP','Selling_Price'])
```

```python
#23. concat & merge:
a=df[:500]
b=df[500:]
c=pd.concat([a,b])
print(c)
```

```python
dummy=pd.DataFrame({'Product Name':['apple','samsung','oneplus'],
                    'Country_of_origin':['India','Japan','us']})
print(pd.merge(df,dummy,on='Product Name'))
```

```python
df['Launch_Date'] = pd.date_range(start='2023-01-01',periods=len(df),freq='D')
print(df['Launch_Date'] )
```

```python
#24. shift:
df['Previous_Row'] = df['Selling_Price'].shift(1)
df.head()
```

```python
#Part 7: Time Series & Numpy
```

```python
#25. Date Creation:
df['Launch_Date']=pd.date_range('2023-01-01',periods=len(df))
df.head()
```

```python
#26. dt Accessors:
df['Year'] = df['Launch_Date'].dt.year
df['Month'] = df['Launch_Date'].dt.month_name()
df['Day'] = df['Launch_Date'].dt.day_name()
df.head(2)
```

```python
#27. Time Periods:
df['Launch_Date']=df['Launch_Date'].dt.to_period('M')
df.head(2)
```

```python
#28. Numpy Integration:
df['Selling_Price'].to_numpy()
percentile= np.percentile(df['Selling_Price'], 75)
print(percentile)
```
